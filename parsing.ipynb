```python
import re
import glob
import csv

manual_borders = []

for filename in sorted(glob.glob("/content/*.txt")):
  with open(filename, 'r', encoding="ISO-8859-1") as f:
    data = f.read()
  find_floats = re.findall(r"[-+]?(?:\d*\.\d+|\d+)", data)
  float_findall = [float(el) for el in find_floats]
  manual_borders.append(float_findall)

print(*manual_borders, sep='\n')
```
